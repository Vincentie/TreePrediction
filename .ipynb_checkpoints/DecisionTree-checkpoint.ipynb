# 以下为训练与调参过程

```python
import numpy as np
import pandas as pd
import sqlite3
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_recall_curve  
from sklearn.metrics import classification_report  
```

```python
#以下为数据处理--构建新指标，以便训练
pd.options.mode.chained_assignment = None

con = sqlite3.connect('FAvgData.sqlite')
sql = "SELECT * from SHFEAvg"
data = pd.read_sql(sql, con, index_col = 'date')
data['close_after5day'] = None
data['max_of_30day'] = None
data['min_of_30day'] = None

for i in range(len(data)-30):
    data['close_after5day'][i] = data['Close'][i+5]
    #5交易日后的价格
    data['max_of_30day'][i+30] = max(data['High'][i:i+30])
    #找出近30交易日内的最大收盘均价。
    data['min_of_30day'][i+30] = min(data['Low'][i:i+30])
    #找出近30交易日内最小收盘均价。

data = data.dropna()   #去掉前无效的数据。
lables_raw = data['close_after5day'] #提取出需要预测的数据
lables = data['close_after5day'] > data['Close'] #为分类处理数据，判断5交易日后的收盘价是否大于今日收盘价

features = data.drop(['close_after5day'],axis = 1) #在特征值中去掉我们要预测的数据。
features = features.drop(['High'],axis = 1) #已有30_max--去掉High
features = features.drop(['Low'],axis = 1)  #已有30_min--去掉Low
```

```python
#未调参的决策树预测
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(features)
features_scaler = scaler.transform(features)
#两行代码用来标准化数据

X_train,X_test, y_train, y_test = train_test_split(features_scaler, lables, test_size = 0.1, random_state = 0)

clf_tree = DecisionTreeClassifier(random_state=0)
clf_tree.fit(X_train, y_train)
print ("预测准确率为：%0.2f" % (clf_tree.score(X_test, y_test)))
```

```python
#调max_depth
import matplotlib.pyplot as plt
i_list = []
score_list = []
for i in range(1,300,1):
    i=i/1.
    clf_tree = DecisionTreeClassifier(max_depth =i )   #使用SVM分类器来判断涨跌
    clf_tree.fit(X_train, y_train)
    i_list.append(i)
    score_list.append(clf_tree.score(X_test, y_test))

score_list_df =  pd.DataFrame({'max_depth':i_list,'score_list':score_list})
```

```python
score_list_df.plot(x='max_depth' ,y='score_list',title='score change with max_depth')
plt.show()
```

```python
#调min_samples_leaf
i_list = []
score_list = []
X_train
for i in range(1,100,1):
    i=i
    clf_tree = DecisionTreeClassifier(min_samples_leaf  = i ) 
    clf_tree.fit(X_train, y_train)
    i_list.append(i)
    score_list.append( clf_tree.score(X_test, y_test) )

score_list_df =  pd.DataFrame({'min_samples_leaf':i_list,'score_list':score_list})
score_list_df.plot(x='min_samples_leaf' ,y='score_list',title='score change with min_samples_leaf')
plt.show()
```

```python
#调min_samples_split
i_list = []
score_list = []

min_samples  =  range(2,50,1)
for i in min_samples :
    clf_tree = DecisionTreeClassifier(min_samples_split  = i ) 
    clf_tree.fit(X_train, y_train)
    i_list.append(i)
    score_list.append(clf_tree.score(X_test, y_test))

score_list_df =  pd.DataFrame({'min_samples_split':i_list,'score_list':score_list})
score_list_df.plot(x='min_samples_split' ,y='score_list',title='score change with min_samples_split')
plt.show()
```

```python
#调min_weight_fraction_leaf
i_list = []
score_list = []

min_weight  =  [x /1000. for x in range(1,500,1)]
for i in min_weight :
    clf_tree = DecisionTreeClassifier(min_weight_fraction_leaf  = i ) 
    clf_tree.fit(X_train, y_train)
    i_list.append(i)
    score_list.append(clf_tree.score(X_test, y_test))

score_list_df =  pd.DataFrame({'min_weight_fraction_leaf':i_list,'score_list':score_list})
score_list_df.plot(x='min_weight_fraction_leaf' ,y='score_list',title='score change with min_weight')
plt.show()
```

```python
#找最优参数
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import ShuffleSplit

X_train,X_test, y_train, y_test = train_test_split(features_scaler, lables, test_size = 0.1, random_state = 0)

clf_tree = DecisionTreeClassifier()
params = { 'max_depth': [x for x in range(1,100,1)], 'min_samples_leaf': [x for x in range(1,40,1)]} #,'min_weight_fraction_leaf':range(0,100,1)

# cv_sets = ShuffleSplit(X_train.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)

grid = GridSearchCV(clf_tree, params)
grid = grid.fit(X_train, y_train)
print (grid.best_estimator_)
```

```python
#调过后的决策树...
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(features)
features_scaler = scaler.transform(features)
#上面两行代码用来标准化数据

X_train,X_test, y_train, y_test = train_test_split(features_scaler, lables, test_size = 0.1, random_state = 0)

clf_tree = DecisionTreeClassifier(max_depth = 9 , min_samples_leaf=22, min_samples_split=2 )   
clf_tree.fit(X_train, y_train)
print ("预测准确率为：%0.2f" % (clf_tree.score(X_test, y_test)))
```

```python
#找出输入指标在预测中的重要性占比
importances = pd.DataFrame({'features':features.columns,'importance':np.round(clf_tree.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False).set_index('features')
importances.plot.bar()
importances
plt.show()
```
